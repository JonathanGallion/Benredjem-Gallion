```python
import pandas as pd
import numpy as np
import math,glob
from scipy.spatial.distance import pdist,squareform
```

```python
for filename in glob.glob('../SEM/Review_freq/*.txt'):
    print(filename)
    nameparts=filename.split('.')
    parts=nameparts[-2].split('/')
    print(parts)
    temp_df=pd.read_csv(filename, sep='\t',header=0,index_col=0)
    temp_df.drop(list(temp_df)[-1], inplace=True, axis=1)
    
    df_values=temp_df.values
    d = pdist(df_values, 'euclidean')
    distance_values = squareform(d)
    dist_df=pd.DataFrame(distance_values,columns=list(temp_df),index=list(temp_df))
    dist_df.to_csv('../SEM/Distance_matrices/DistanceMatrix_%s.csv'%(parts[-1]))
    
```
